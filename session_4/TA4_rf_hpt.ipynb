#### Handling Categorical Variables

We will be preparing the handling of the categorical variables and also preparing the different sets for the pre-processing step of the pipeline.

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()
cat_cols = df.select_dtypes(include="object").columns.tolist()
print(df.columns)

"""
We establish several sets of features: numerical, categorical, binary and probabilistic
    - Numerical: Features that take continuous values
    - Categorical: Features that represent categories or groups. From those, we will drop 'time_signature' for simplicity.
        - OneHot: We will one-hot encode 'key'
        - TargetEncode: We will target encode 'track_genre'
    - Binary: Features that have two possible values (e.g., 0/1), passed through the pre-processing steps
    - Probabilistic: Features that represent probabilities or likelihoods (e.g. danceability, speechiness), passed through the pre-processing steps
"""

num_cols = ["duration_ms", "tempo", "loudness"]
cat_cols = ["track_genre", "key", "time_signature"]
bin_cols = ["explicit", "major"]
prob_cols = ["danceability", "energy", "speechiness", "acousticness", "instrumentalness", "liveness", "valence"]

# Really the important lists from categorical are these two:
oh_cols = ["key"]
te_cols = ["track_genre"]
```

```python
"""
At a first glance, there are some features which really need some handling: key, time_signature and track_genre (bonus track, explicit).
    - key: This is a categorical variable but it's currently being treated as a numerical variable. I'm really not a fan of the way it's encoded so I'll start by reverting it.
            We will be turning it back into a categorical variable and then you choose whether to one-hot encode it or not.
    - time_signature: This variable is also categorical and should be treated as such. But more important than that, 0/4 and 1/4 are REALLY weird values.
                        We will be turning it into two separate binary variables: 3_4 and 4_4 (5_4 is REALLY weird to dance to so it might be worth having it as the baseline).
    - track_genre: This is another categorical variable that needs to be encoded.
                    You will be choosing whether to drop it or use target encoding on it.
    - explicit: Simply turn it from a True/False binary into a 0/1 binary, no big deal.
"""

# We start with "explicit" because the solution is super simple.
df["explicit"] = df["explicit"].astype(int)

# Now we map the numerical key values to their corresponding musical note names, for clarity if we choose to one-hot encode it later.
key_map = {
    0: "C",
    1: "C#/Db",
    2: "D",
    3: "D#/Eb",
    4: "E",
    5: "F",
    6: "F#/Gb",
    7: "G",
    8: "G#/Ab",
    9: "A",
    10: "A#/Bb",
    11: "B"
}

df["key"] = df["key"].map(key_map)

df
```

```python
"""
The features which are still defined in categories are: (unique)track_id, artists, album_name, track_name and track_genre.

Out of those, we can only realistically use track_genre. Any other of those variables is more of an identification variable.
Any predictive power that could come from those variables would probably have it because of being from a specific genre or from another one.
An artist who usually produces edm might have most of their tracks as highly danceable, but if they suddenly release an album of piano ballads, that could change
for that album specifically. Granted, this change in danceability would be captured by the album variable, but it would also be captured by the track_genre variable anyway.

To sum up, we are only keeping track_genre for reasons of avoiding having unmanageable dimensionality and to prevent the model to rely too much on features it can't be trained on
when making predictions for unseen data (newly released tracks).
"""

df = df.drop(columns=["track_id", "artists", "album_name", "track_name"])
df
```

#### Preprocessing Step using Column Transformer

- One hot encoder: It's not necessary to integrate it in the pipeline, but it's cleaner if we do since we are already using sklearn pipelines.

- Target Encoder: In this case it IS essential to integrate target encoder in the pipeline to avoid data leakage. If we encode using the target outside of the pipeline we would need a very careful cross-validation scheme to avoid data leakage. By integrating it in the pipeline we ensure that the encoding is done only on the training folds during cross-validation, thus avoiding data leakage.

```python
onehot_enc = OneHotEncoder(sparse_output=False, drop="first")
target_encoder = TargetEncoder(target_type="continuous", random_state=SEED)

preproc = ColumnTransformer(
        transformers=[
            ("num", StandardScaler(), num_cols),
            ("oh", onehot_enc, oh_cols),
            ("te", target_encoder, te_cols),
            ("bin", "passthrough", bin_cols),
            ("prob", "passthrough", prob_cols)  
        ],
        remainder="drop"  # This drops any columns not specified in the transformers
)
```

#### Make Pipeline function

It's just a function that we will use to create a pipeline with the chosen estimator below. We switch algorithms by hand in this case, we are not including algorithm selection in the gridsearch to keep things simpler.

```python
"""
Our pipeline is reasonably simple, but if we had a lot of preprocessing, we could define a cached memory for the preprocedding and TTR steps
in order to avoid having to fit them every time. I don't think this is necessary here though.
"""

def make_pipeline(estimator, target_function, inverse_target_function):
    ttr = TransformedTargetRegressor(
        regressor=estimator,
        func=target_function,
        inverse_func=inverse_target_function
    )
    
    return Pipeline([
        ("preproc", preproc),
        ("model", ttr)
    ])
```

```python
"""
We simply predict the target as it is [0-100],
but for the inverse, we simply clip the predicted values to be within [0, 100].
"""

def y_to_t(y):
    t = np.sqrt(y/100)
    return t

def t_to_y(y):
    t = y**2 * 100
    return np.clip(t, 0, 100)
```

## Optuna Hyperparameter Tuning

You could use OptunaSearchCV, which integrates nicely as a direct replacement for GridSearchCV, but you miss out on some cool features like conditional spaces or multiple model families in the same study.

As you can see, I'm using:
- Conditional spaces: To avoid errors with combinations of bootstrap=False and max_samples being set (max_samples only allows "None" when bootstrap=False).
- Multiple model families: To try out both Random Forest and Elastic Net in the same study.

That being said, you can find a simpler example using OptunaSearchCV below.

```python
"""
Full Optuna implementation
"""

# We define the objective function for Optuna

# But first, we need to create a Cross-Validation structure ourselves here, because optuna doesn't do it for us
# We set the cv here because we want to compare the same folds for each trial
# (although setting a random_state in KFold should be enough for that, but I set it outside as a profilactic measure in case I forget to set random_state)
cv = KFold(n_splits=5, shuffle=True, random_state=SEED)

def objective(trial):
    # Hyperparameter search space
    
    # Alternatives for model selection
    model_type = trial.suggest_categorical("model_type", ["rf", "en"]) 
    
    # Conditional HPs based on model type selected
    # Conditional HPs for Random Forest
    if model_type == "rf":
        # HP ranges to optimize
        n_estimators = trial.suggest_int("rf_n_estimators", 50, 500)
        max_depth = trial.suggest_int("rf_max_depth", 10, 100)
        min_samples_split = trial.suggest_int("rf_min_samples_split", 2, 10, step=2)
        min_samples_leaf = trial.suggest_int("rf_min_samples_leaf", 1, 10)
        max_features = trial.suggest_categorical("rf_max_features", ["sqrt", None])
        bootstrap = trial.suggest_categorical("rf_bootstrap", [False, True])
        
        # max_samples doesn't properly work when bootstrap=False (needs to be set to None, specifically)
        # plus there is no benefit at all in trying max_samples=None when bootstrap=True either
        if bootstrap:
            max_samples = trial.suggest_float("rf_max_samples", 0.5, 1.0)
        else:
            max_samples = None
        
        # We initialize the RF object and create the pipeline using it
        rf = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            max_features=max_features,
            bootstrap=bootstrap,
            max_samples=max_samples,
            n_jobs=-1, # I set n_jobs=-1 here to speed up the training, because I have 8 cores, so if I use n_jobs=-1 in cv, I'm not using all of them
            random_state=SEED
        )
        
        model = make_pipeline(rf, y_to_t, t_to_y)
    
    # Conditional parameters for Elastic Net
    elif model_type == "en":
        # HP ranges
        alpha = trial.suggest_float("en_alpha", 0.01, 10.0, log=True)
        l1_ratio = trial.suggest_float("en_l1_ratio", 0.0, 1.0)
        fit_intercept = trial.suggest_categorical("en_fit_intercept", [True, False])
        
        # EN object and pipeline
        en = ElasticNet(
            alpha=alpha,
            l1_ratio=l1_ratio,
            fit_intercept=fit_intercept,
            random_state=SEED
        )
        
        model = make_pipeline(en, y_to_t, t_to_y)
    
    # Simply for mental completeness, although optuna should never suggest an invalid model_type
    # I added it as a safeguard in case I messed up in some naming, so that the error tells me exactly where to look
    else:
        raise ValueError("Invalid model type")
    
    time_start = time.time()
    cv_metrics = cross_val_score(
        model,
        X_train,
        y_train,
        cv=cv,
        scoring="neg_mean_squared_error", # option one, where then we set direction="maximize" in the Optuna study or we multiply by -1 later, I'll do that for now
        # scoring=make_scorer(mean_squared_error, greater_is_better=False), # option two, where we keep direction="minimize"
        n_jobs=1, # Would use only 5 out of 8 cores in my machine so I set n_jobs=1 here to use all cores for each fold in the algorithm
    )
    time_end = time.time()
    
    mean_cv_mse = -np.mean(cv_metrics)  # because cross_val_score returns negative MSE for neg_mean_squared_error
    time_taken = time_end - time_start
    
    """
    The main advantage of specifying the score in this way (inside the objective function itself) is that
    we can specifically control how we want the trade-off between metrics to be handled. 
    
    For classification tasks, we could combine accuracy and recall in a single score,
    but give more weight to accuracy or more weight to recall depending on the use case.
    
    Giving multiple metrics to Optuna to optimize uses a different sampler and a defined Pareto front
    to decide which combination constitutes a "better" score. TPE is generally regarded as the best sampler from
    Optuna so I prefer to use a single metric and control how it's computed. Nevertheless, you need to be very careful
    when doing this because you can easily end up with a model that overfits to a specific metric and performs poorly.
    
    In this case, I show you how to combine MSE with the time taken to perform cros-validation,
    but I would not recommend you actually doing this unless you have a very specific reason to do so.
    Probably a better combination would be MSE with MAE or with R2 score, giving some residual weight to either MAE or R2
    (i.e. keeping MSE as the main metrics to optimize). this implementation would be trivial given the structure already shown.
    """
    
    score = mean_cv_mse # + time_taken * 1e-3  # Adding a small penalty for longer training times if needed
    return score
```

```python
db_path = Path("../data/spotify/optuna_spotify.db").resolve()

study = optuna.create_study(
    sampler=optuna.samplers.TPESampler(seed=SEED),
    direction="minimize", # Super important to tweak this when using MSE (lower is better so we minimize, unlike with accuracy)
    storage=f"sqlite:///{db_path.as_posix()}",
    study_name="spotify_model_optimization",
    load_if_exists=True
)

study.optimize(objective, n_trials=50, show_progress_bar=True)

print("Best params:")
print(study.best_params)
print(f"Best MSE: {study.best_value:.2f}")
```

## Model Evaluation

#### Baseline Model

Given that our target variable takes values between 0 and 100, we can create a baseline for comparison, to evaluate how good our model is.

To do this, we will compute the RMSE of a naive model that always predicts the mean of the target variable in the training set.

```python
def mean_prediction(X_test, y_train):
    """
    This function computes the mean of the target variable y and returns it as predictions for all observations in X_test.
    """
    y_bar = np.mean(y_train)
    return np.ones(X_test.shape[0]) * y_bar

y_pred_mean = mean_prediction(X_test, y_train)
print("Mean prediction RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_mean)))
```
